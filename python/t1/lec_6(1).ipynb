```python
#  faltu

import pandas as pd
import plotly.express as px

# Sample data
data = {
    'hours' : [2,3,4,5,6],
    'marks' : [40,50,60,70,80],
    'attendance' : [60,65,70,80,90],
    'day' : ['Mon','Tue','Wed','Thu','Fri']  # Added for animation
}
df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(
    df,
    x='hours',
    y='marks',
    z='attendance',
    color='marks',            # Color by marks
    size='attendance',        # Size by attendance
    symbol='marks',           # Different symbols for marks
    animation_frame='day',    # Animate across days
    hover_data=['hours','marks','attendance','day']  # Extra info on hover
)

# Style upgrades
fig.update_layout(
    template='plotly_dark',   # Dark futuristic theme
    scene=dict(
        xaxis=dict(backgroundcolor="black", gridcolor="gray"),
        yaxis=dict(backgroundcolor="black", gridcolor="gray"),
        zaxis=dict(backgroundcolor="black", gridcolor="gray")
    ),
    title="Cool Interactive 3D Scatter Plot"
)

fig.show()
```

```python
data = {
    'hours' : [2,3,4,5,6],
    'marks' : [40,50,60,70,80],
    'attendance' : [60,65,70,80,90]
    }

df = pd.DataFrame(data)

fig = px.scatter_matrix(
    df ,
    dimensions = ['hours' , 'marks' , 'attendance'],
    color='marks',
    title='scatter',
    symbol='marks'
)
fig.show()
```
