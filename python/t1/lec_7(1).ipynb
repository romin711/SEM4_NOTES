```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = {
    'department' : ['it' , 'hr' , 'finance' , 'marketing'],
    'employees' : [40 , 20 , 25 , 15]
 }

df = pd.DataFrame(data)
fig = px.pie(df , values='employees' , names='department' , color_discrete_sequence=['blue' , 'green' , 'orange' , 'red'] , title='employee distribution')
fig.update_traces(textinfo='percent+label' , textposition='inside' , pull=[0,0,0.1,0])
```

```python
data = {
    'marks' : [40,45,50,55,60,65,70,75,80,85],
    'gender' : ['male' , 'female' , 'male' , 'female' , 'male' , 'female' , 'male' , 'female' , 'male' , 'female']
}

df = pd.DataFrame(data)
fig = px.histogram(df , x='marks' , nbins=10 , color='gender' , histfunc='count' , pattern_shape='gender' , text_auto=True , color_discrete_map={'male' : 'blue' ,'female' : 'pink' })
fig.show()
```
